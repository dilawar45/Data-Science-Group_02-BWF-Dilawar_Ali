```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
exercise = sns.load_dataset('exercise')
exercise.head()
```

# **1. Scatter Plot**

```python
sns.scatterplot(x='pulse',y='time',data=exercise)
plt.title('Pulse vs Time')
plt.xlabel('Pulse')
plt.ylabel('Time')
plt.show()
```

# **2. Line Plot**

```python
# Line plot
sns.lineplot(data=exercise, x="time", y="pulse",hue='kind')
plt.title("Line Plot of Pulse Over Time by Kind")
plt.xlabel("Time")
plt.ylabel("Pulse")
plt.show()
```

# **Bar Plot**

```python
# Bar plot
sns.barplot(data=exercise, x="diet", y="pulse", hue="kind")
plt.title("Bar Plot of Average Pulse by Diet and Kind")
plt.xlabel("Diet")
plt.ylabel("Average Pulse")
plt.show()
```

# **Count Plot**

```python
sns.countplot(data=exercise, x="diet")
plt.title("Count Plot of Diet")
plt.xlabel("Diet")
plt.ylabel("Count")
plt.show()
```

# **Box Plot**

```python
sns.boxplot(data=exercise, x="diet", y="pulse")
plt.title("Box Plot of Pulse by Diet")
plt.xlabel("Diet")
plt.ylabel("Pulse")
plt.show()
```

# **Setting the Style use: sns.set_style()**


```python
sns.set_style("darkgrid")
sns.scatterplot(x='pulse',y='time',data=exercise)
plt.title('Pulse vs Time')
plt.xlabel('Pulse')
plt.ylabel('Time')
plt.show()
```
