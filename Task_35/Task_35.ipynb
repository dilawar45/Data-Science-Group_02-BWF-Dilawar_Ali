```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
path='/content/drive/MyDrive/ByteWise Tasks/Image.jpg'
```

```python
img = Image.open(path)

# Convert to numpy array
img_np = np.array(img)
```

**Function to Display Image**

```python
def Display(img, title, cmap=None):
    plt.figure(figsize=(6, 4))
    if cmap is None:
        if len(img.shape) == 2:  # Grayscale image
            cmap = 'gray'
        else:  # Color image (RGB)
            cmap = None
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
Display(img_np, 'Original Image')
```

**Applying Different Filters**

```python
gray = cv2.cvtColor(img_np, cv2.COLOR_RGB2GRAY)
Display(gray, 'Grayscale')
```

```python
# Gaussian Blur
gaussian_blur = cv2.GaussianBlur(img_np, (15, 15), 0)
Display(gaussian_blur, 'Gaussian Blur')
```

```python
# Gaussian Blur
gaussian_blur = cv2.GaussianBlur(img_np, (15, 15), 0)
Display(gaussian_blur, 'Gaussian Blur')
```

```python
# Sharpening Filter
kernel = np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]])
sharpened = cv2.filter2D(img_np, -1, kernel)
Display(sharpened, 'Sharpening Filter')
```

```python
# Detail Enhancement
detail_enhanced = cv2.detailEnhance(img_np, sigma_s=10, sigma_r=0.15)
Display(detail_enhanced, 'Detail Enhancement')
```

```python
# Median Filter
median_filtered = cv2.medianBlur(img_np, 5)
Display(median_filtered, 'Median Filter')
```

```python
# Inversion
inverted = cv2.bitwise_not(img_np)
Display(inverted, 'Inversion')
```
