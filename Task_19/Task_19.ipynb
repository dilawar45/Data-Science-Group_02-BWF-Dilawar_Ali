```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Using Built in dataset
diamond=sns.load_dataset('diamonds')
diamond.head()
```

# **Linear Regression Plot**

```python
# Linear Regression Plot
sns.regplot(data=diamond,x='carat',y='price',)
plt.title("Linear Regression Plot")
plt.show()
```

# **Multiple Regression**

```python
sns.lmplot(data=diamond,x='carat',y='price',hue='cut')
plt.title("Multiple Regression Plot")
plt.show()
```

# **Residual Plot**

```python
sns.residplot(data=diamond,x='carat',y='price')
plt.title("Residual Plot")
plt.show()
```

# **Histogram**

```python
sns.histplot(data=diamond,x='price')
plt.title("Histogram Plot")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

# **Kernel Density Estimate (KDE) Plot**

```python
sns.kdeplot(data=diamond,x='price',fill=True)
plt.title("Kernel Density Plot")
plt.xlabel("Price")
plt.ylabel("Density")
plt.show()
```

# **Distribution Plot**

```python
sns.histplot(data=diamond,x='price',bins=30,kde=True)
plt.title("Distribution Plot")
plt.xlabel("Price")
plt.ylabel("Frequency/Density")
plt.show()
```

# **Box Plot**

```python
# Box plot
sns.boxplot(data=diamond,x="cut", y="price")
plt.title("Box Plot of Diamond Prices by Cut")
plt.xlabel("Cut")
plt.ylabel("Price")
plt.show()
```

# **Violin Plot**

```python
# Violin plot
sns.violinplot(data=diamond,x="cut", y="price")
plt.title("Violin Plot of Diamond Prices by Cut")
plt.xlabel("Cut")
plt.ylabel("Price")
plt.show()
```

# **Pair Plot**

```python
# Pair plot
sns.pairplot(data=diamond,vars=["carat", "price", "depth", "table"], hue="cut")
plt.suptitle("Pair Plot of Diamonds Dataset", y=1.02)
plt.show()
```

# **HeatMap**

```python
# Exclude non-numerical columns
numerical_data = diamond.select_dtypes(include='number')
# Calculate the correlation matrix
corr = numerical_data.corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Heatmap of Diamond Features Correlation")
plt.show()
```

# **Joint Plot**

```python
# Joint plot
sns.jointplot(data=diamond, x="carat", y="price", kind="reg")
plt.suptitle("Joint Plot of Diamond Carat and Price", y=1.02)
plt.show()
```

# **Facet Grid**

```python
# Facet Grid
g = sns.FacetGrid(data=diamond, col="cut")
g.map(sns.scatterplot, "carat", "price")
plt.suptitle("Facet Grid of Diamond Carat and Price by Cut", y=1.02)
plt.show()
```
