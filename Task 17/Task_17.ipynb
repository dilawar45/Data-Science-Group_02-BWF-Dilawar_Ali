```python
path='/content/drive/MyDrive/university_enrollment_(Mini-project-01).csv'
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv(path)
data.head()
```

# **Subplots**

**Generate a figure with 2x2 subplots, each containing a different type of plot (e.g., line plot, scatter plot, bar plot, and histogram).**


```python
# Prepare subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Line plot for Enrollment Trends Over the Years
enrollment_trends = data.groupby('year')['enrollment_count'].sum()
axs[0, 0].plot(enrollment_trends.index, enrollment_trends.values, marker='o', color='b')
axs[0, 0].set_title('Enrollment Trends Over the Years')
axs[0, 0].set_xlabel('Year')
axs[0, 0].set_ylabel('Total Enrollment')
axs[0, 0].grid(True)

# Scatter plot for Pre-Scores vs Post-Scores
axs[0, 1].scatter(data['pre_score'], data['post_score'], alpha=0.5, color='r')
axs[0, 1].set_title('Pre-Scores vs Post-Scores')
axs[0, 1].set_xlabel('Pre-Score')
axs[0, 1].set_ylabel('Post-Score')
axs[0, 1].grid(True)

# Bar plot for Average Enrollment by Course Type
average_enrollment = data.groupby('course_type')['enrollment_count'].mean()
axs[1, 0].bar(average_enrollment.index, average_enrollment.values, color=['g', 'c'])
axs[1, 0].set_title('Average Enrollment by Course Type')
axs[1, 0].set_xlabel('Course Type')
axs[1, 0].set_ylabel('Average Enrollment')
axs[1, 0].grid(True)

# Histogram for Enrollment Count
axs[1, 1].hist(data['enrollment_count'], bins=10, color='m', alpha=0.7)
axs[1, 1].set_title('Histogram of Enrollment Count')
axs[1, 1].set_xlabel('Enrollment Count')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()

```

# **Historgam**

```python
# Histogram for Enrollment Count with 10 bins
plt.hist(data['enrollment_count'], bins=10, alpha=0.5, label='Enrollment Count', color='b')

plt.title('Histogram of Enrollment Count')
plt.xlabel('Enrollment Count')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

# Normalized Histograms
plt.hist(data['enrollment_count'], bins=10, alpha=0.5, label='Enrollment Count', color='b', density=True)

plt.title('Normalized Histogram of Enrollment Count')
plt.xlabel('Enrollment Count')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

# **Box Plot**

```python
# Box plot for Enrollment Count by Department
plt.figure(figsize=(10, 6))
sns.boxplot(x='department', y='enrollment_count', data=data, palette='Set2')
plt.title('Box Plot of Enrollment Count by Department')
plt.xlabel('Department')
plt.ylabel('Enrollment Count')
plt.grid(True)
plt.show()

# Multiple box plots in a single figure
plt.figure(figsize=(14, 8))
sns.boxplot(x='department', y='enrollment_count', data=data, palette='Set2')
plt.title('Box Plots of Enrollment Count by Department')
plt.xlabel('Department')
plt.ylabel('Enrollment Count')
plt.grid(True)
plt.show()
```
