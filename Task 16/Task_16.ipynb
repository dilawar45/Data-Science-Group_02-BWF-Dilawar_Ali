```python
import matplotlib.pyplot as plt
import numpy as np
```

# **Q.1 Basic Plots**

```python
#Simple line plot
x=np.array([0,6])
y=np.array([0,50])
plt.plot(x,y)
plt.show()
```

```python
# Data for the first line
x1 = np.array([0, 6])
y1 = np.array([0, 50])

# Data for the second line
x2 = np.array([0, 6])
y2 = np.array([0, 30])

# Plot the first line
plt.plot(x1, y1)

# Plot the second line
plt.plot(x2, y2)

# Show the plot
plt.show()
```

```python
# Adding marker
# Data for the first line
x1 = np.array([0, 6])
y1 = np.array([0, 50])

# Data for the second line
x2 = np.array([0, 6])
y2 = np.array([0, 30])

# Plot the first line
plt.plot(x1, y1, marker='o')

# Plot the second line
plt.plot(x2, y2, marker='d')

# Show the plot
plt.show()
```

```python
# Customize style and color of lines
# Data for the first line
x1 = np.array([0, 6])
y1 = np.array([0, 50])

# Data for the second line
x2 = np.array([0, 6])
y2 = np.array([0, 30])

# Plot the first line
plt.plot(x1, y1, marker='o',ls='--',c='r')

# Plot the second line
plt.plot(x2, y2, marker='d',ls='-.',c='g')

# Show the plot
plt.show()
```

# **Q.2  Scatter Plots:**

```python
# Basic Satter Plot
x=np.random.rand(40)
y=np.random.rand(40)
plt.scatter(x,y)
plt.show()
```

```python
#Customize Marker Styles, Sizes, and Colors
x=np.random.rand(40)
y=np.random.rand(40)
plt.scatter(x,y,marker='o',s=100,c='r')
plt.show()
```

```python
# Plot Multiple Scatter Plots on the Same Graph
x=np.random.rand(25)
y=np.random.rand(25)
x2 = np.random.rand(25)
y2 = np.random.rand(25)

plt.scatter(x, y, marker='o', s=100, c='blue')
plt.scatter(x2, y2, marker='x', s=100, c='green')
plt.show()
```

```python
# Adding labels and titles
plt.scatter(x, y, marker='o', s=100, c='blue', label='Dataset 1')
plt.scatter(x2, y2, marker='x', s=100, c='green', label='Dataset 2')
plt.title("Scatter Plot with Labels and Titles")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

# **Q.3 Bar Charts**

```python
#Create a vertical bar chart.
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()


```

```python
# Create a horizontal bar chart.
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x,y)
plt.show()
```

```python
# Customize bar colors and add edge colors.

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color='orange', edgecolor='black')
plt.show()
```

# **Q.4 Pie Charts**

```python
#Create a basic pie chart.
y = np.array([35, 25, 25, 15])
plt.pie(y)
plt.show()
```

```python
#Customize slice colors and add a legend.
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
#Explode a slice of the pie chart.
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels,explode=[0.1,0,0,0])
```

```python
#Add percentage labels to slices.
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels,autopct='%1.1f%%')
```
