```python
path='/content/drive/MyDrive/university_enrollment_(Mini-project-01).csv'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**1. Average Enrollment for Online vs Classroom Courses:**

```python
data = pd.read_csv(path)
average_enrollment = data.groupby('course_type')['enrollment_count'].mean()
print("Average Enrollment for Online vs Classroom Courses:\n", average_enrollment)
```

**2. Enrollment Trends Over the Years for Online Courses**

```python
online_courses = data[data['course_type'] == 'online']
online_trends = online_courses.groupby('year')['enrollment_count'].sum()
plt.plot(online_trends.index, online_trends.values, marker='o')
plt.title('Online Course Enrollment Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Enrollment')
plt.grid(True)
plt.show()
```

**3. Department with the Highest Average Enrollment**

```python
average_enrollment_department = data.groupby('department')['enrollment_count'].mean().idxmax()
print("Department with the Highest Average Enrollment:", average_enrollment_department)
```

**4. Correlation Between Pre-Scores and Post-Scores**

```python
data['pre_score'] = pd.to_numeric(data['pre_score'], errors='coerce')
data['post_score'] = pd.to_numeric(data['post_score'], errors='coerce')

clean_data = data.dropna(subset=['pre_score', 'post_score'])
correlation = clean_data['pre_score'].corr(clean_data['post_score'])
print(f"Correlation coefficient between pre-scores and post-scores: {correlation}")

# Plot correlation
sns.scatterplot(x='pre_score', y='post_score', data=clean_data)
plt.title('Correlation Between Pre-Scores and Post-Scores')
plt.xlabel('Pre-Score')
plt.ylabel('Post-Score')
plt.grid(True)
plt.show()
```

**5. Beginner Courses vs Intermediate Courses Enrollment**

```python
beginner_courses = data[data['pre_requirement'] == 'Beginner']['enrollment_count']
intermediate_courses = data[data['pre_requirement'] == 'Intermediate']['enrollment_count']
average_beginner_enrollment = beginner_courses.mean()
average_intermediate_enrollment = intermediate_courses.mean()
print(f"Average Beginner Course Enrollment: {average_beginner_enrollment}")
print(f"Average Intermediate Course Enrollment: {average_intermediate_enrollment}")
```

**6. Average Post-Score for Online vs Classroom Courses**



```python
average_post_score = data.groupby('course_type')['post_score'].mean()
print("Average Post-Score for Online vs Classroom Courses:\n", average_post_score)
```

**7. Enrollment Trends Over the Years**

```python
yearly_enrollment = data.groupby('year')['enrollment_count'].sum()
plt.plot(yearly_enrollment.index, yearly_enrollment.values, marker='o')
plt.title('Total Enrollment Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Enrollment')
plt.grid(True)
plt.show()
```

**8. Distribution of Scores Across Departments**

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='department', y='pre_score', data=clean_data)
plt.title('Distribution of Pre-Scores Across Departments')
plt.xlabel('Department')
plt.ylabel('Pre-Score')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='department', y='post_score', data=clean_data)
plt.title('Distribution of Post-Scores Across Departments')
plt.xlabel('Department')
plt.ylabel('Post-Score')
plt.grid(True)
plt.show()
```

**9. Year with the Highest Total Enrollment**

```python
year_with_highest_enrollment = yearly_enrollment.idxmax()
print(f"Year with the Highest Total Enrollment: {year_with_highest_enrollment}")
```

**10. Average Enrollment for Courses with vs without Pre-Requirement**

```python
with_pre_requirement = data[data['pre_requirement'].notna()]['enrollment_count']
without_pre_requirement = data[data['pre_requirement'].isna()]['enrollment_count']
average_with_pre_requirement = with_pre_requirement.mean()
average_without_pre_requirement = without_pre_requirement.mean()
print(f"Average Enrollment with Pre-Requirement: {average_with_pre_requirement}")
print(f"Average Enrollment without Pre-Requirement: {average_without_pre_requirement}")
```

**11. Average Improvement (Post-Score - Pre-Score) for Each Course Type**

```python
data['improvement'] = data['post_score'] - data['pre_score']
average_improvement = data.groupby('course_type')['improvement'].mean()
print("Average Improvement for Each Course Type:\n", average_improvement)
```

**12. Course Type with Higher Average Post-Score**

```python
higher_average_post_score_course_type = data.groupby('course_type')['post_score'].mean().idxmax()
print("Course Type with Higher Average Post-Score:", higher_average_post_score_course_type)
```

**13. Departments More Likely to Offer Online Courses**

```python
online_departments = data[data['course_type'] == 'online']['department'].value_counts(normalize=True)
print("Departments More Likely to Offer Online Courses:\n", online_departments)
```

**14. Median Enrollment for Courses Over the Years**

```python
median_enrollment_by_year = data.groupby('year')['enrollment_count'].median()
print("Median Enrollment for Courses Over the Years:\n", median_enrollment_by_year)
```

**15. Significant Difference in Enrollment Counts Between Online and Classroom Courses**

```python
significant_difference_enrollment = data.groupby('course_type')['enrollment_count'].mean()
print("Significant Difference in Enrollment Counts Between Online and Classroom Courses:\n", significant_difference_enrollment)
```

**16. Courses with the Highest Enrollment Count**

```python
highest_enrollment_courses = data[data['enrollment_count'] == data['enrollment_count'].max()]
print("Courses with the Highest Enrollment Count:\n", highest_enrollment_courses)
```

**17. Pre-Scores for Online vs Classroom Courses**

```python
average_pre_score = data.groupby('course_type')['pre_score'].mean()
print("Average Pre-Scores for Online vs Classroom Courses:\n", average_pre_score)
```

**18. Overall Enrollment Trend for Each Department Over the Years**

```python
department_enrollment_trends = data.groupby(['year', 'department'])['enrollment_count'].sum().unstack()
department_enrollment_trends.plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Overall Enrollment Trend for Each Department Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Enrollment')
plt.grid(True)
plt.legend(title='Department')
plt.show()
```

**9. Significant Difference in Post-Scores Based on Course Type**

```python
significant_difference_post_scores = data.groupby('course_type')['post_score'].mean()
print("Significant Difference in Post-Scores Based on Course Type:\n", significant_difference_post_scores)
```

**20. Course Type Showing More Improvement from Pre-Score to Post-Score**

```python
course_type_improvement = data.groupby('course_type')['improvement'].mean()
print("Course Type Showing More Improvement from Pre-Score to Post-Score:\n", course_type_improvement)
```
