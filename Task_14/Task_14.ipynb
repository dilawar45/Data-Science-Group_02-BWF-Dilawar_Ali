```python
import pandas as pd
```

#### Identify missing values in the DataFrame.

```python
df.isnull().sum()
```

#### Drop rows with any missing values.

```python
df_dropna_rows = df.dropna()
df_dropna_rows
```

#### Drop columns with any missing values.

```python
df_dropna_cols = df.dropna(axis=1)
df_dropna_cols
```

#### Fill missing values with a specific value.

```python
df_filled = df.fillna(0)
df_filled
```

```python
df_filled.isnull().sum()
```

#### Fill missing values using forward fill and backward fill methods.

```python
df_ffill = df.fillna(method='ffill')
df_ffill
```

```python
df_bfill = df.fillna(method='bfill')
df_bfill
```

#### Interpolate missing values.

```python
df_interpolated = df.interpolate()
df_interpolated
```

#### Convert a column to a different data type.

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Age'] = df['Age'].astype(int)
```

```python
df.info()
```

#### Apply a function to transform the values of a column.

```python
def convert_gender(gender):
    if gender.lower() == 'male':
        return 'm'
    elif gender.lower() == 'female':
        return 'f'
    return gender
```

```python
df['Sex'] = df['Sex'].apply(convert_gender)
```

#### Normalize a column using Min-Max scaling.

```python
df['Fare_normalized'] = (df['Fare'] - df['Fare'].min()) / (df['Fare'].max() - df['Fare'].min())
df
```

#### Standardize a column (z-score normalization).

```python
df['Fare_scaled'] = (df['Fare'] - df['Fare'].mean()) / df['Fare'].std()
df
```

#### Identify duplicate rows in the DataFrame.

```python
duplicates = df.duplicated()
print(duplicates.sum())
```

#### Drop duplicate rows.

```python
df_no_duplicates = df.drop_duplicates()
```

#### Drop duplicate rows based on specific columns.

```python
df_no_duplicates_subset = df.drop_duplicates(subset=['Age', 'Sex'])
```

#### Convert all string values in a column to lowercase.

```python
df['Name'] = df['Name'].str.lower()
df
```

#### Remove leading and trailing spaces from string values in a column.

```python
df['Name'] = df['Name'].str.strip()
df.head()
```

#### Replace a specific substring in a column with another substring.

```python
df['Sex'] = df['Sex'].str.replace('female', 'F')
df.head()
```

#### Extract a substring from each value in a column.

```python
df['cabin_section'] = df['Cabin'].str[:1]
df.head()
```

#### Convert a column to datetime format.

```python
df['synthetic_date'] = pd.to_datetime('2024-01-01')
df.head()
```

#### Extract year, month, and day from a datetime column.

```python
df['year'] = df['synthetic_date'].dt.year
df['month'] = df['synthetic_date'].dt.month
df['day'] = df['synthetic_date'].dt.day
df.head()
```

#### Filter rows based on a date range.

```python
date_filtered = df[(df['synthetic_date'] > '2023-01-01') & (df['synthetic_date'] < '2024-12-31')]
date_filtered.head()
```

#### Group values in a categorical column and create a new column with grouped categories.

```python
df['age_group'] = pd.cut(df['Age'], bins=[0, 18, 40, 60, 80], labels=['Child', 'Adult', 'Middle-aged', 'Senior'])
df.head()
```

#### Merge two DataFrames based on a common column.

```python
df_additional = pd.DataFrame({'Name': df['Name'], 'ticket_class': df['Pclass']})
merged_df = pd.merge(df, df_additional, on='Name')
merged_df.head()
```

#### Concatenate two DataFrames vertically.

```python
concatenated_df = pd.concat([df, df])
concatenated_df.head()
```

#### Concatenate two DataFrames horizontally.

```python
concatenated_df_horizontal = pd.concat([df, df_additional], axis=1)
concatenated_df_horizontal.head()
```

#### Create a new column based on existing columns.

```python
df['family_size'] = df['SibSp'] + df['Parch']
df.head()
```

#### Discretize a continuous column into bins.

```python
df['Age_binned'] = pd.cut(df['Age'], bins=5, labels=False)
df[['Age', 'Age_binned']]
```
