**Task 13: Data manipulation with Pandas (indexing, selection, grouping)**

```python
import pandas as pd
```

**1. Load a DataFrame from a CSV file. Display the first and last five rows of the DataFrame.**

```python
df = pd.read_csv("/content/drive/MyDrive/covid_19_clean_complete.csv")
df
```

```python
df.head()
```

```python
df.tail()
```

**2. Set a specific column as the index of the DataFrame.**

```python
df.set_index('Country/Region', inplace=True)
df.head()
```

**3. Select a specific column and display its values.**

```python
confirmed_values = df['Confirmed']
confirmed_values
```

**4. Select multiple columns and display the resulting DataFrame.**

```python
selected_columns_df = df[['Confirmed', 'Deaths']]
selected_columns_df
```

**5. Select a subset of rows using the .loc method.**

```python
subset_loc = df.loc['Afghanistan']
subset_loc
```

**6. Select a subset of rows and columns using the .iloc method.**

```python
subset_iloc = df.iloc[0:5, 0:3]
subset_iloc
```

**7. Filter rows based on a condition.**

```python
filtered_rows = df[df['Confirmed'] > 0]
filtered_rows
```

**8. Group the DataFrame by a specific column and calculate the mean of each group.**


```python
mean_by_country = df.groupby('Country/Region').mean(numeric_only=True)
mean_by_country
```

**9. Group the DataFrame by multiple columns and calculate the sum of each group.**

```python
sum_by_country_region = df.groupby(['Country/Region']).sum(numeric_only=True)
sum_by_country_region
```

**10. Use the agg method to apply multiple aggregation functions to grouped data.**

```python
agg_functions = df.groupby('Country/Region').agg({'Confirmed': ['mean', 'max'], 'Deaths': ['mean', 'min']})
agg_functions
```

**11. Calculate the size of each group.**

```python
size_by_country = df.groupby('Country/Region').size()
size_by_country
```

**12. Select rows based on multiple conditions.**

```python
multiple_conditions = df[(df['Confirmed'] > 1000) & (df['Deaths'] < 100)]
multiple_conditions
```

**13. Use the query method to filter rows.**

```python
query_filtered = df.query('Confirmed > 1000')
query_filtered
```
