
# **Task 34-> Deep Learning with TensorFlow/Keras**

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/ByteWise Tasks/file.csv')
df.head()
```

```python
# Feature selection
features = ['Dew Point Temp_C', 'Rel Hum_%', 'Wind Speed_km/h', 'Visibility_km', 'Press_kPa']
X = df[features]
# Target variable
y = df['Temp_C']
# Normalize the features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the model
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1)  # Output layer for regression (predicting 'Temp_C')
])
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
model.fit(X_train, y_train, epochs=100, validation_split=0.2, batch_size=32)
```

```python
test_loss, test_mae = model.evaluate(X_test, y_test)
print('\nTest MAE:', test_mae)
```
